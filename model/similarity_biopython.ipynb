```python
from Bio import pairwise2
from Bio.Seq import Seq

import pandas as pd
```

```python
peptides = pd.read_csv('amp+avp (seq+label).csv')

peptides
```

```python
for i in peptides['sequence']:
    print(i)
```

```python
def calculate_sequence_similarity(seq1, seq2):
    # Convert the sequences to Seq objects
    seq1 = Seq(seq1)
    seq2 = Seq(seq2)

    # Perform global sequence alignment
    alignments = pairwise2.align.globalxx(seq1, seq2)

    # Get the alignment with the best score (the first one in the list)
    best_alignment = alignments[0]
    seq1_aligned, seq2_aligned, score, begin, end = best_alignment

    # Calculate similarity as a percentage
    similarity = (score / max(len(seq1), len(seq2))) * 100

    return similarity
```

```python
similarity = calculate_sequence_similarity("KMGRLFR", "KAGRSFR")
print(f"Similarity between the sequences: {similarity:.2f}%")
```

```python
similarity_scores = []

for i in peptides['sequence']:
    row_scores = []
    for j in peptides['sequence']:
        similarity = calculate_sequence_similarity(i, j)
        print("The similarity between peptides", i, "and", j, "is: {}".format(similarity))
        row_scores.append(similarity)
    similarity_scores.append(row_scores)
```
