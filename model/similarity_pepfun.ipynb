```python
from pepfun.sequence import Sequence
from pepfun.sequence import Library

from pepfun.extra import Alignment
from pepfun.extra import pepDescriptors

# Additional functions to convert SMILES to peptide
from pepfun.sequence import Sequence
from pepfun.extra import readProperties
from pepfun.extra import peptideFromSMILES

import pandas as pd
```

```python
peptides = pd.read_csv('amp+avp (seq+label).csv')

peptides
```

```python
for i in peptides['sequence']:
    print(i)
```

```python
similarity = Alignment.similarity_pair("KMGRLFR", "KAGRSFR")
print("The similarity between peptides KMGRLFR and KAGRSFR is: {}".format(similarity))
```

```python
similarity_scores = []

for i in peptides['sequence']:
    row_scores = []
    for j in peptides['sequence']:
        similarity = Alignment.similarity_pair(i, j)
        print("The similarity between peptides", i, "and", j, "is: {}".format(similarity))
        row_scores.append(similarity)
    similarity_scores.append(row_scores)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a heatmap plot
sns.heatmap(similarity_scores, cmap='coolwarm', annot=True, vmin=-1, vmax=1)
plt.xlabel('Peptides')
plt.ylabel('Peptides')
plt.title('Peptide Similarity')
plt.show()
```
